# Classical decomposition

```python
from statsmodels.tsa.seasonal import seasonal_decompose
import matplotlib as mpl

mpl.rc("figure", figsize=(12,6))
result_add_train = seasonal_decompose(data['value'], model='additive', period = 12)

result_add_train.plot();
```

```python
result_mul = seasonal_decompose(data['value'], model='multiplicative',period = 6)

result_mul.plot();
```

### เรียกดูแต่ละส่วน

```python
print(result_add_train.trend)
print(result_add_train.seasonal)
print(result_add_train.resid)
print(result_add_train.observed)
```

### ปฏิบัติการครั้งที่ 7
1. สร้างโมเดลเพื่อการทำนายดังนี้
1.1 แบ่งข้อมูลออกเป็น 2 ส่วน
   - training set: วันที่ 1996-01-01 ถึง 2009-12-01
   - test set: วันที่ 2010-01-01 ถึง 2012-03-01  
1.2 แยกส่วนประกอบ $y_t=S_t+T_t+R_t$ บน training set
1.3 สร้าง time series ชุดใหม่ที่แสดงถึงทำนายค่าบน test set โดยนำค่า $T_t+R_t$ ของวันล่าสุดใน training set ที่มีค่า $T_t$ มาบวกกับแต่ละค่าใน $S_t$ จากวันที่ วันที่ 2010-01-01 ถึง 2012-03-01  
3. คำนวณ RMSE โดยใช้ฟังก์ชัน `rmse` ข้างล่าง
4. แสดงแผนภาพข้อมูล elecequip และค่าทำนายที่ได้

```python
training_set = data.loc['1996-01-01':'2009-12-01']
test_set = data.loc['2010-01-01':'2012-03-01']
```

```python
result_add = seasonal_decompose(training_set['value'], model='additive', period = 12)
```

```python
result_add_test = seasonal_decompose(test_set['value'], model='additive', period = 12)
```

```python
trend = result_add.trend.dropna()[-1]
residual = result_add.resid.dropna()[-1]
```

```python
constant = trend + residual
constant
```

```python
ts_new = result_add_test.seasonal + constant
ts_new
```

```python
evalution = rmse(test_set['value'],ts_new)
evalution
```

```python
def rmse(y_true,y_pred):
  # Inputs
  # y_true: actual value, y_pred: predicted values
  # Output: RMSE between y_true and y_pred
  return np.sqrt(np.mean(np.square(y_true-y_pred)))
```
